### Introduction

The importance of mental health in college students cannot be emphasized enough. As students leave everything familiar to them and enter university, they often experience emotional and mental strain. Students from different courses, ages, and years may encounter mental breakdowns at some point in their lives. A key aspect of mental health issues among students is depression.

**Depression**

According to Wikipedia, depression is classified medically as a mental and behavioral disorder. The experience of depression affects a person's thoughts, behavior, motivation, feelings, and sense of well-being. Symptoms of depression include:

- Consistent feelings of hopelessness and sadness
- Mood swings
- Changes in sleep and/or appetite
- Withdrawal from social circles, a tendency to self-isolate
- Increased pessimism
- Feeling lethargic
- Difficulty concentrating and completing tasks
- Lack of enjoyment in activities one previously found pleasurable

This dataset was collected by the author through a survey conducted via Google Forms from university students to examine their current academic situation and mental health. The data is based on Malaysia and was collected from IIUM (International Islamic University Malaysia).

The dataset is publicly available on Kaggle [here](https://www.kaggle.com/datasets/shariful07/student-mental-health).

In this project, we focus on predicting depression by providing student information. By analyzing various factors and indicators related to students' mental health, we aim to identify those at risk of depression and provide insights for early intervention and support.

```python
# dropping Timestamp from dataset
new_df.drop('Timestamp', axis=1, inplace=True)
```

```python
# Applying oridinal encoding to CGPA
cgpa_mapping = {
    '0 - 1.99': 1,
    '2.00 - 2.49': 2,
    '2.50 - 2.99': 3,
    '3.00 - 3.49': 4,
    '3.50 - 4.00': 5
}
new_df['What is your CGPA?'] = new_df['What is your CGPA?'].replace(cgpa_mapping)
new_df['What is your CGPA?']
```

```python
# Non oneHotEcoded data
non_one_hot_encoded = new_df.drop(one_hot_columns, axis=1)
non_one_hot_encoded
```

```python
# Applying label encoding to dependent variable
label_encoder = LabelEncoder()
non_one_hot_encoded['Do you have Depression?'] = label_encoder.fit_transform(non_one_hot_encoded['Do you have Depression?'])
```

```python
# Marging one-hot-encoded dataframe with non-one-hot-encoded dataframe
encoded_data = pd.concat([one_hot_encoded_df,non_one_hot_encoded], axis=1)
encoded_data
```

```python
X = encoded_data.drop('Do you have Depression?', axis=1)
y = encoded_data['Do you have Depression?']
```

```python
# Combine the scaled features and target into a single dataframe
df_combined_scaled = pd.concat([pd.DataFrame(X_scaled, columns=X.columns), y.reset_index(drop=True)], axis=1)
```

```python
# find unique vlaues of y varibale
unique, counts = np.unique(y, return_counts=True)
print(unique)
print(counts)
```

####  Determining n_componets for ML Models

```python
# Determining n_componets for ML Models
pca = PCA()
principal_components = pca.fit_transform(X)
explained_variance_FD = pca.explained_variance_ratio_
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

A scree plot in Principal Component Analysis (PCA) is a visual tool used to determine the optimal number of principal components to retain. In this graph, the x-axis represents the principal components, while the y-axis shows the variance explained by each component. The plot displays a characteristic steep decline in the first few components, followed by a more gradual decrease, and finally a long tail where additional components contribute minimal explained variance. Based on this scree plot, choosing around 21 principal components appears reasonable as there's a noticeable "elbow" around this point, after which the curve flattens significantly. This choice strikes a balance between dimensionality reduction and information retention, likely capturing a substantial portion of the data's variance while eliminating less important features. However, the final decision should also consider other factors such as cumulative explained variance and the specific requirements of the analysis at hand.

```python
import numpy as np
cumulative_variance = np.cumsum(explained_variance_FD)
threshold = [0.5,0.7,0.8,0.9]
for i in threshold:
    n_components = np.argmax(cumulative_variance >= i) + 1
    print(f"n components (threshold {i}%) : {n_components}")
```

Based on the analysis of the data using Principal Component Analysis (PCA), considering 28 components that explain 70% of the variance appears to be a reasonable approach. This choice strikes a balance between significant dimensionality reduction and retention of crucial information. By using cumulative variance, we can quantitatively assess how much of the total data variation is captured by a given number of components, providing a more objective basis for selection compared to visual inspection of the scree plot alone. Cumulative variance is particularly useful because it allows us to set specific thresholds (like 70% in this case) and identify the number of components needed to reach that threshold. This method ensures we retain the most important structures in the data while substantially reducing its complexity. Opting for 28 components offers a middle ground between the more aggressive reduction suggested by the scree plot's elbow (around 20 components) and the higher number (36) required to explain 80% of the variance and (43) required to explain 90% of the variance. This compromise likely captures the most significant patterns in the data while making subsequent analyses more manageable and computationally efficient.

```python
pca = PCA(n_components=28)
X_pca = pca.fit_transform(X)
X_pca
```

# ML Models

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y.values, test_size=0.3, random_state=42)
```

##Training Models

### Logistic Regression
Initialize, train, and evaluate the Logistic Regression model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Initialize model
log_reg = LogisticRegression()

# Fit model
log_reg.fit(X_train, y_train)

# Predict probabilities
y_prob_log_reg = log_reg.predict_proba(X_test)[:, 1]

# Predict labels
y_pred_log_reg = log_reg.predict(X_test)

# Compute ROC curve
fpr_log_reg, tpr_log_reg, _ = roc_curve(y_test, y_prob_log_reg)

# Compute AUC
roc_auc_log_reg = auc(fpr_log_reg, tpr_log_reg)

# Plot ROC curve for Logistic Regression
plt.figure()
plt.plot(fpr_log_reg, tpr_log_reg, color='blue', lw=2, label='Logistic Regression (AUC = %0.2f)' % roc_auc_log_reg)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Logistic Regression')
plt.legend(loc="lower right")
plt.show()

# Print AUC result
print(f'Logistic Regression AUC: {roc_auc_log_reg:.2f}')

# Generate and print classification report
class_report_log_reg = classification_report(y_test, y_pred_log_reg)
print('Classification Report for Logistic Regression:')
print(class_report_log_reg)

# Compute confusion matrix
conf_matrix_log_reg = confusion_matrix(y_test, y_pred_log_reg)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_log_reg)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix - Logistic Regression')
plt.show()
```

### Random Forest
Initialize, train, and evaluate the Random Forest model.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Initialize model
rand_forest = RandomForestClassifier()

# Fit model
rand_forest.fit(X_train, y_train)

# Predict probabilities
y_prob_rand_forest = rand_forest.predict_proba(X_test)[:, 1]

# Predict labels
y_pred_rand_forest = rand_forest.predict(X_test)

# Compute ROC curve
fpr_rand_forest, tpr_rand_forest, _ = roc_curve(y_test, y_prob_rand_forest)

# Compute AUC
roc_auc_rand_forest = auc(fpr_rand_forest, tpr_rand_forest)

# Plot ROC curve for Random Forest
plt.figure()
plt.plot(fpr_rand_forest, tpr_rand_forest, color='red', lw=2, label='Random Forest (AUC = %0.2f)' % roc_auc_rand_forest)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Random Forest')
plt.legend(loc="lower right")
plt.show()

# Print AUC result
print(f'Random Forest AUC: {roc_auc_rand_forest:.2f}')

# Generate and print classification report
class_report_rand_forest = classification_report(y_test, y_pred_rand_forest)
print('Classification Report for Random Forest:')
print(class_report_rand_forest)

# Compute confusion matrix
conf_matrix_rand_forest = confusion_matrix(y_test, y_pred_rand_forest)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_rand_forest)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix - Random Forest')
plt.show()
```

### Naive Bayes
Initialize, train, and evaluate the Naive Bayes model.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_curve, auc, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Initialize model
naive_bayes = GaussianNB()

# Fit model
naive_bayes.fit(X_train, y_train)

# Predict probabilities
y_prob_naive_bayes = naive_bayes.predict_proba(X_test)[:, 1]

# Predict labels
y_pred_naive_bayes = naive_bayes.predict(X_test)

# Compute ROC curve
fpr_naive_bayes, tpr_naive_bayes, _ = roc_curve(y_test, y_prob_naive_bayes)

# Compute AUC
roc_auc_naive_bayes = auc(fpr_naive_bayes, tpr_naive_bayes)

# Plot ROC curve for Naive Bayes
plt.figure()
plt.plot(fpr_naive_bayes, tpr_naive_bayes, color='orange', lw=2, label='Naive Bayes (AUC = %0.2f)' % roc_auc_naive_bayes)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Naive Bayes')
plt.legend(loc="lower right")
plt.show()

# Print AUC result
print(f'Naive Bayes AUC: {roc_auc_naive_bayes:.2f}')

# Generate and print classification report
class_report_naive_bayes = classification_report(y_test, y_pred_naive_bayes)
print('Classification Report for Naive Bayes:')
print(class_report_naive_bayes)

# Compute confusion matrix
conf_matrix_naive_bayes = confusion_matrix(y_test, y_pred_naive_bayes)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_naive_bayes)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix - Naive Bayes')
plt.show()
```

### SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Initialize model with probability=True to enable probability predictions
svm = SVC(probability=True)

# Fit model
svm.fit(X_train, y_train)

# Predict probabilities
y_prob_svm = svm.predict_proba(X_test)[:, 1]

# Predict labels
y_pred_svm = svm.predict(X_test)

# Compute ROC curve
fpr_svm, tpr_svm, _ = roc_curve(y_test, y_prob_svm)

# Compute AUC
roc_auc_svm = auc(fpr_svm, tpr_svm)

# Plot ROC curve for SVM
plt.figure()
plt.plot(fpr_svm, tpr_svm, color='purple', lw=2, label='SVM (AUC = %0.2f)' % roc_auc_svm)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - SVM')
plt.legend(loc="lower right")
plt.show()

# Print AUC result
print(f'SVM AUC: {roc_auc_svm:.2f}')

# Generate and print classification report
class_report_svm = classification_report(y_test, y_pred_svm)
print('Classification Report for SVM:')
print(class_report_svm)

# Compute confusion matrix
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_svm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix - SVM')
plt.show()
```

###Hyperparameter Tuning for SVM

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Define the parameter grid
param_grid_svm = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'poly', 'sigmoid']
}

# Initialize GridSearchCV with SVM
grid_search_svm = GridSearchCV(estimator=SVC(probability=True), param_grid=param_grid_svm, cv=5, scoring='roc_auc', n_jobs=-1)

# Fit the model with the best hyperparameters
grid_search_svm.fit(X_train, y_train)

# Get the best model
best_svm = grid_search_svm.best_estimator_

# Evaluate the best model
y_prob_best_svm = best_svm.predict_proba(X_test)[:, 1]
y_pred_best_svm = best_svm.predict(X_test)
fpr_best_svm, tpr_best_svm, _ = roc_curve(y_test, y_prob_best_svm)
roc_auc_best_svm = auc(fpr_best_svm, tpr_best_svm)

# Print results
print(f'Best SVM AUC: {roc_auc_best_svm:.2f}')
print(f'Best Hyperparameters for SVM: {grid_search_svm.best_params_}')
```

```python
# Compute ROC curve for the best SVM model
fpr_best_svm, tpr_best_svm, _ = roc_curve(y_test, y_prob_best_svm)
roc_auc_best_svm = auc(fpr_best_svm, tpr_best_svm)

# Plot ROC curve for the best SVM model
plt.figure()
plt.plot(fpr_best_svm, tpr_best_svm, color='purple', lw=2, label='Best SVM (AUC = %0.2f)' % roc_auc_best_svm)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Best SVM')
plt.legend(loc="lower right")
plt.show()

# Print AUC result
print(f'Best SVM AUC: {roc_auc_best_svm:.2f}')
```
